```python
import pandas as pd # type: ignore
import numpy as np # type: ignore
```

```python
data = pd.read_csv(r"big-data-change2-identifyingdataremoved.csv")
```

```python
zero_quantities: dict[int, set[str]] = {}
for index in range(2096):
    zeros = sum([1 for x in data.iloc[index] if x == 0])
    if zeros not in zero_quantities: zero_quantities[zeros] = set()
    zero_quantities[zeros].add(index)
for zero_count in sorted(zero_quantities):
    print(f"{zero_count}: {zero_quantities[zero_count]} -> {round(zero_count / 63 * 100, 3)}%")
```

```python
zero_quantities: dict[int, set[str]] = {}
for x in data:
    zeros = sum([1 for y in data[x] if y == 0])
    if zeros not in zero_quantities: zero_quantities[zeros] = set()
    zero_quantities[zeros].add(x)
for zero_count in sorted(zero_quantities):
    print(f"{zero_count}: {zero_quantities[zero_count]} -> {round(zero_count / 2096 * 100, 3)}%")
```
